```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing


pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.float_format = '{:,.2f}'.format
```

```python
data = pd.read_csv('/content/houses_Madrid.csv', on_bad_lines='skip') # skip lines that cause errors
```

```python
data.info()
```



*   Remove nan column
*   Take specific column for Processing

```python
#Remove nan column
data =data.dropna( axis=1, how='all' )
data.info()
```

```python
#check if there dublicate value
data.duplicated().sum()
```

```python
data.rename(columns = {'subtitle': 'city'}, inplace = True)
```

```python
df=data[['id','city','sq_mt_built','n_rooms','n_bathrooms','rent_price','buy_price','buy_price_by_area']]
df.info()
```

```python
df.n_rooms.describe()
```

```python
df.drop(df[df['n_rooms'] == 0].index, inplace = True)
```

```python
df.n_bathrooms.mode()
```

```python
df.n_bathrooms=df.n_bathrooms.fillna(1)
df.n_bathrooms=df.n_bathrooms.replace(0,1)
df.n_bathrooms=df.n_bathrooms.astype(int)
```

```python
df.info()
```

```python
df =df.dropna(subset=['rent_price', 'sq_mt_built'])
```

```python
df.sample(10)
```

```python
#change negative value
for value in df.rent_price.values:
  if value < 0:
    df.rent_price = 0
```

```python
#split cigty to (disrect,city)
df['destrict']=df.city.str.split(',').str[0]
```

```python
df= df[['id', 'destrict', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'rent_price',
       'buy_price', 'buy_price_by_area']]
df.head()
```
